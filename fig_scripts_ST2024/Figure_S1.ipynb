```python
def calc_gregory(TS_sim, FLNT_sim, FSNT_sim, TS_con, FLNT_con, FSNT_con):
    TS_diff = TS_sim-TS_con
    FSNT_diff = FSNT_sim-FSNT_con
    FLNT_diff = FLNT_sim-FLNT_con
    TOA_diff = FSNT_diff-FLNT_diff
    m_rapid_FLNT, b_rapid_FLNT = np.polyfit(TS_diff[:15], FLNT_diff[:15],1)
    m_slow_FLNT, b_slow_FLNT = np.polyfit(TS_diff[50:], FLNT_diff[50:],1)
    m_rapid_FSNT, b_rapid_FSNT = np.polyfit(TS_diff[:15], FSNT_diff[:15],1)
    m_slow_FSNT, b_slow_FSNT = np.polyfit(TS_diff[50:], FSNT_diff[50:],1)
    No = b_rapid_FSNT-b_rapid_FLNT
    lam_rapid = m_rapid_FSNT-m_rapid_FLNT
    lam_slow = m_slow_FSNT-m_slow_FLNT
    b_slow_net = b_slow_FSNT-b_slow_FLNT
    b_rapid_net = b_rapid_FSNT-b_rapid_FLNT
    delta_Ts = np.round(-b_slow_net/lam_slow,2)
    return(No, delta_Ts, lam_rapid, lam_slow, m_rapid_FSNT, m_rapid_FLNT, m_slow_FSNT, m_slow_FLNT)
```

```python
No_8xco2_nostrato, delta_Ts_8xco2_nostrato, lam_rapid_8xco2_nostrato, lam_slow_8xco2_nostrato, m_rapid_FSNT_8xco2_nostrato, m_rapid_FLNT_8xco2_nostrato, m_slow_FSNT_8xco2_nostrato, m_slow_FLNT_8xco2_nostrato = calc_gregory(TS_8xco2_nostrato_greg, FLNT_8xco2_nostrato_greg, FSNT_8xco2_nostrato_greg, np.average(TS_8xco2_greg[-20:]), np.average(FLNT_8xco2_greg[-20:]), np.average(FSNT_8xco2_greg[-20:]))
No_8xco2_strato, delta_Ts_8xco2_strato, lam_rapid_8xco2_strato, lam_slow_8xco2_strato, m_rapid_FSNT_8xco2_strato, m_rapid_FLNT_8xco2_strato, m_slow_FSNT_8xco2_strato, m_slow_FLNT_8xco2_strato = calc_gregory(TS_8xco2_strato_greg, FLNT_8xco2_strato_greg, FSNT_8xco2_strato_greg, np.average(TS_8xco2_greg[-20:]), np.average(FLNT_8xco2_greg[-20:]), np.average(FSNT_8xco2_greg[-20:]))
No_1xco2_nostrato, delta_Ts_1xco2_nostrato, lam_rapid_1xco2_nostrato, lam_slow_1xco2_nostrato, m_rapid_FSNT_1xco2_nostrato, m_rapid_FLNT_1xco2_nostrato, m_slow_FSNT_1xco2_nostrato, m_slow_FLNT_1xco2_nostrato = calc_gregory(TS_1xco2_nostrato_greg, FLNT_1xco2_nostrato_greg, FSNT_1xco2_nostrato_greg, np.average(TS_1xco2_greg[-20:]), np.average(FLNT_1xco2_greg[-20:]), np.average(FSNT_1xco2_greg[-20:]))
No_1xco2_strato, delta_Ts_1xco2_strato, lam_rapid_1xco2_strato, lam_slow_1xco2_strato, m_rapid_FSNT_1xco2_strato, m_rapid_FLNT_1xco2_strato, m_slow_FSNT_1xco2_strato, m_slow_FLNT_1xco2_strato = calc_gregory(TS_1xco2_strato_greg, FLNT_1xco2_strato_greg, FSNT_1xco2_strato_greg, np.average(TS_1xco2_greg[-20:]), np.average(FLNT_1xco2_greg[-20:]), np.average(FSNT_1xco2_greg[-20:]))
```
