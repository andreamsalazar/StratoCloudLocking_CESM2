```python
def get_700hPa_lev(x, year_start, year_end): # input surface pressure
    z700_x =  np.logical_and(1000*hyam[:, None, None] +np.inner(hybm[:, None, None],np.reshape(np.mean(x[year_start:year_end,:,:], axis=0),(96,144,1)))[:,0,:,:]>70000, 1000*hyam[:, None, None] +np.inner(hybm[:, None, None],np.reshape(np.mean(x[year_start:year_end,:,:], axis=0),(96,144,1)))[:,0,:,:]<75000)
    z700_x[z700_x==False]=0
    z700_x[z700_x==True]=1
    z700_x_arg = np.argmax(z700_x, axis = 0)
    return(z700_x_arg) # outputs level of around 700hPa
```
